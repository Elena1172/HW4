```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('kc-house-data.csv')
df.date = pd.to_datetime(df.date)
df.head(3)
```

###### 1.2 Изучите стоимости недвижимости

```python
plt.rc('xtick',labelsize=10)
plt.rc('ytick',labelsize=10)
plt.figure(figsize=(6, 3))
plt.hist(df['price'], bins=40)
plt.title('Стоимость домов, проданных за период 2014 - 2015 гг.')
plt.xlabel('стоимость домов')
plt.ylabel('количество домов');
```

```python
plt.figure(figsize=(8, 3))
sns.histplot(x = df.price, bins=40)
plt.title('Распределение стоимости домов')
plt.xlabel('Стоимость домов')
plt.ylabel('Количество');
```

```python
plt.figure(figsize=(5, 2))
sns.boxplot(x=df.price)
plt.title('Ящик с усами для цен')
plt.xlabel('Стоимость домов в млн.')
plt.ylabel('Количество домов');
```

Больше всего домов со стоимостью менее 1 000 000. Есть и выбросы: дома со стоимостью около 8 000 000

###### 2.3 Изучите распределение состояния домов

```python
cor_mat=np.round(df.corr(),1)
cor_mat[np.abs(cor_mat)<0.4]=0
cor_mat.loc['price']
```

+ Коэффициент корреляции показывает что существует линейная зависимость между стоимостью домов и количеством ванных комнат.Зависимость прямая средняя (заметная) (corr=0.5).
+ Стоимость домов зависит также от размера жилой площади и состоянием. Зависимость прямая сильная(corr=0.7).
+ Зависимость от вида из окна прямая слабая (cor=0.4).
+ Зависимость от размера участка площади над землей  и жилой площади соседей. Зависимость прямая средняя (corr=0.6)


```python
plt.figure(figsize=(12, 4))
sns.boxplot(x=df['bathrooms'], y=df['price'])
plt.title('Зависимость стоимости дома от количества ванных комнат')
plt.xlabel('количество ванных комнат')
plt.ylabel('стоимость дома');
```

Диаграмма показывает как растет стоимость дома с изменением количества ванных комнат. Боксплоты по количествам ванных комнат. Почти везде стоимость дома увеличивается с увеличением количества ванных комнат.

```python
plt.figure(figsize=(12, 4))
sns.scatterplot(x=df.sqft_living, y=df.price)
plt.title('Диаграмма изменения цены в зависимости от размера жилой площади дома')
plt.xlabel('Жилая площадь дома')
plt.ylabel('Стоимость дома в млн.');
```

На диаграмме хорошо видно как растет цена с увеличением жилой площади

```python
plt.figure(figsize=(12, 4))
sns.boxplot(x=df.grade, y=df.price)
plt.title('Зависимость стоимости дома от состояния')
plt.xlabel('Состояние дома (учитывается конструкция и дизайн)')
plt.ylabel('Стоимость дома в млн.');
```

Данные показывают что стоимость дома напрямую зависит от его состояния

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df.sqft_above, y=df.price)
plt.title('Изменение стоимости дома в зависимости от изменения площади над землей')
plt.xlabel('Площадь дома на поверхности земли')
plt.ylabel('Стоимость дома в млн');
```

Расположение данных показывает, что стоимость дома увеличивается с увеличением площади дома на поверхности 

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x=df.sqft_living15, y=df.price)
plt.title('Диаграмма изменения стоимости домов в зависимости от изменения общей жилой площади у 15 соседей')
plt.xlabel('Общая жилая площадь у 15 соседей')
plt.ylabel('Стоимость домов в млн.');
```
